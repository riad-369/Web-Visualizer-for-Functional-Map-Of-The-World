```python
#! /bin/python3
# Count categories in the dataset for CSE4510
# Written By Tyler Zars
import pandas as pd
import ast

groundtruth_data = pd.read_csv("data/groundtruthdata.csv")

# Get All Rows Categories List
categorical_items = []
counter = 0

# For each row in the set, add the category to the list
for img in groundtruth_data[['bounding_boxes']].values:
    correct_list = ast.literal_eval(img[0])
    #print(res[0]['raw_category'])
    categorical_items.append(correct_list[0]['raw_category'])
    
    # Print as it goes through (3000 for more accurate)
    if counter % 20000 == 0:
        print(correct_list[0]['raw_category'])
    
    #break
    counter += 1

# Add New Row
groundtruth_data['category'] = pd.Series(categorical_items)
```

```python
#! /bin/python3
from shapely.geometry import LineString
import re

def center_point(polygon_obj):
    # POLYGON ((12.757161911800001 47.270718547500003, 12.7616534882 47.270718547500003, 12.7616534882 47.273775608000001, 12.757161911800001 47.273775608000001, 12.757161911800001 47.270718547500003))

    # Initial break
    split_by_comma = polygon_obj.split(",")

    # Break insides up
    split_by_spaces = []
    for chunk in split_by_comma:
        split_by_spaces.append(chunk.split(" "))

    # Points list for computing
    points_list = []

    for item in split_by_spaces: 
        indv_points_list = []
        for position in range(len(item)):
            if position != 0:
                # Remove anything thats not the number
                to_append = re.findall("\d+(?:\.\d+)?", item[position])
                indv_points_list.append(to_append)

        # Tuples    
        tuple_fun = (float(indv_points_list[0][0]), float(indv_points_list[1][0]))
            

        points_list.append(tuple_fun)
    
    # See final list of [[point], [point], ...]
    print(points_list)

    # Makes shapely object for centroid function
    lines = LineString(points_list)
    center = lines.centroid
    return (center.x, center.y)

tasty = center_point("POLYGON ((-12.757161911800001 47.270718547500003, -12.7616534882 47.270718547500003, -12.7616534882 47.273775608000001, -12.757161911800001 47.273775608000001, -12.757161911800001 47.270718547500003))")
print(tasty)
```

```python
npp_geo = groundtruth_data
npp_geo['lat'] = npp_geo.apply(lambda row: center_point(row['raw_location'])[1], axis=1)
npp_geo['long'] = npp_geo.apply(lambda row: center_point(row['raw_location'])[0], axis=1)
npp_geo = geopandas.GeoDataFrame(npp_geo, geometry=geopandas.points_from_xy(npp_geo['long'], npp_geo['lat']))
npp_geo
```
